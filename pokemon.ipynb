```python
missing_height = df[df['height_m'].isnull()]
missing_weight = df[df['weight_kg'].isnull()]

# Check for overlap
overlap = missing_height.index.intersection(missing_weight.index)
print(f"Number of rows missing both: {len(overlap)}")
```

```python
# Drop rows with missing values in height_m and weight_kg
df = df.dropna(subset=['height_m', 'weight_kg'])
df
```

```python
# Filling with 50% as default assumption
#df['percentage_male'].fillna(50, inplace=True)
df.fillna({'percentage_male': 50}, inplace=True) #or df[col] = df[col].method(value)
```

```python
# Filling missing values with "None"
#df['type2'].fillna('None', inplace=True)
df.fillna({'type2': 'None'}, inplace=True)
df['type2']
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates if found
#if duplicates > 0:
 #   pokemon.drop_duplicates(inplace=True)
```

```python
# Convert capture_rate to numeric
df['capture_rate'] = pd.to_numeric(df['capture_rate'], errors='coerce')
```

```python
# Convert object type columns to category
columns_to_category = ['type1', 'type2', 'classfication', 'abilities']
for col in columns_to_category:
    df[col] = df[col].astype('category')
```

```python
# Check for remaining missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Drop the row with missing value in capture_rate
df = df[df['capture_rate'].notnull()]
print(df.isnull().sum())
```

# Univariate Analysis

```python
# This column shows Pokémon's overall stats. Let’s visualize and calculate statistics.

# Plot histogram and KDE for base_total
plt.figure(figsize=(10, 6))
sns.histplot(df['base_total'], kde=True, bins=20, color='blue')
plt.title('Distribution of Base Total')
plt.xlabel('Base Total')
plt.ylabel('Frequency')
plt.show()

# Summary statistics for base_total
base_total_stats = df['base_total'].describe().round(2)
print(base_total_stats)
```

```python
# Plot histogram and KDE for weight_kg
plt.figure(figsize=(10, 6))
sns.histplot(df['weight_kg'].dropna(), kde=True, bins=20, color='green')
plt.title('Distribution of Weight (kg)')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()

# Plot histogram and KDE for height_m
plt.figure(figsize=(10, 6))
sns.histplot(df['height_m'].dropna(), kde=True, bins=20, color='orange')
plt.title('Distribution of Height (m)')
plt.xlabel('Height (m)')
plt.ylabel('Frequency')
plt.show()

# Summary statistics
weight_stats = df['weight_kg'].describe().round(2)
height_stats = df['height_m'].describe().round(2)
print("Weight Statistics:\n", weight_stats)
print("\nHeight Statistics:\n", height_stats)
```

```python
# Frequency of type1
type1_counts = df['type1'].value_counts()

# Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=type1_counts.index, y=type1_counts.values, palette='viridis')
plt.title('Frequency of Primary Types (type1)')
plt.xlabel('Type1')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Print frequency table
print(type1_counts)
```

```python
# Frequency of generation
generation_counts = df['generation'].value_counts()

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=generation_counts.index, y=generation_counts.values, palette='coolwarm')
plt.title('Frequency of Pokémon by Generation')
plt.xlabel('Generation')
plt.ylabel('Count')
plt.show()

# Print frequency table
print(generation_counts)
```

```python
# Separate legendary and non-legendary
legendary_stats = df[df['is_legendary'] == 1]['base_total']
non_legendary_stats = df[df['is_legendary'] == 0]['base_total']

# Box plot comparison
plt.figure(figsize=(10, 6))
sns.boxplot(x='is_legendary', y='base_total', data=df, palette='Set2')
plt.title('Comparison of Base Total: Legendary vs. Non-Legendary Pokémon')
plt.xlabel('Is Legendary (0 = No, 1 = Yes)')
plt.ylabel('Base Total')
plt.show()

# Print statistics
print("Legendary Base Total Stats:\n", legendary_stats.describe().round(2))
print("\nNon-Legendary Base Total Stats:\n", non_legendary_stats.describe().round(2))
```

```python
import math
from collections import Counter

# Split abilities and count occurrences
abilities = df['abilities'].str.strip("[]").str.replace("'", "").str.split(", ")
abilities_counts = Counter([item for sublist in abilities.dropna() for item in sublist])

# Determine the number of subplots needed
chunk_size = 46  # Number of abilities per subplot
abilities_chunks = [list(abilities_counts.items())[i:i+chunk_size] for i in range(0, len(abilities_counts), chunk_size)]

# Calculate the number of rows for the subplots
n_rows = math.ceil(len(abilities_chunks))

# Create subplots
fig, axes = plt.subplots(n_rows, 1, figsize=(12, 6 * n_rows), constrained_layout=True)

# If there's only one row, convert axes to a list for consistency
if n_rows == 1:
    axes = [axes]

# Plot each chunk in a separate subplot
for i, (ax, chunk) in enumerate(zip(axes, abilities_chunks)):
    abilities, counts = zip(*chunk)
    sns.barplot(x=list(abilities), y=list(counts), ax=ax, palette='cubehelix')
    ax.set_title(f'Frequency of Abilities (Chunk {i+1})', fontsize=14)
    ax.set_xlabel('Ability', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.tick_params(axis='x', rotation=60, labelsize=10)

plt.show()
```

```python
from collections import Counter

# Specify the number of top abilities to plot
top_n = 10

# Extract the top abilities using most_common
top_abilities = Counter(abilities_counts).most_common(top_n)

# Convert to separate lists for plotting
abilities, counts = zip(*top_abilities)

# Plot the top abilities
plt.figure(figsize=(12, 6))
sns.barplot(x=list(abilities), y=list(counts), palette="viridis")
plt.title(f"Top {top_n} Most Frequent Abilities", fontsize=14)
plt.xlabel("Ability", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# All Stats Comparison
# To get an overall picture, compare attack, defense, speed, and base_total side-by-side.

# Combine stats for a grouped box plot
stats_to_compare = ['attack', 'defense', 'speed', 'base_total']
df_melted = df.melt(id_vars='is_legendary', value_vars=stats_to_compare, 
                    var_name='Stat', value_name='Value')

# Box plot for all stats
plt.figure(figsize=(14, 8))
sns.boxplot(x='Stat', y='Value', hue='is_legendary', data=df_melted, palette='pastel')
plt.title('Comparison of Stats: Legendary vs. Non-Legendary Pokémon')
plt.xlabel('Stat')
plt.ylabel('Value')
plt.legend(title='Is Legendary', labels=['Non-Legendary', 'Legendary'])
plt.show()
```

```python
# Define the stats to compare
stats_to_compare = ['attack', 'defense', 'speed', 'base_total']

# Melt the DataFrame for a grouped boxplot
df_melted = df.melt(
    id_vars='is_legendary',
    value_vars=stats_to_compare,
    var_name='Stat',
    value_name='Value'
)

# Box plot for all stats
plt.figure(figsize=(14, 8))
sns.boxplot(
    x='Stat',
    y='Value',
    hue='is_legendary',
    data=df_melted,
    palette={0: 'blue', 1: 'orange'}
)

# Customizing plot titles and axes
plt.title('Comparison of Stats: Legendary vs Non-Legendary Pokémon', fontsize=16)
plt.xlabel('Stat', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Fixing the legend with clear labels
legend_handles = [
    plt.Line2D([0], [0], color='blue', lw=4, label='Non-Legendary'),
    plt.Line2D([0], [0], color='orange', lw=4, label='Legendary')
]
plt.legend(handles=legend_handles, title='Legendary Status')

# Display the plot
plt.show()
```
